```python
biomart_mapping = pd.read_csv("mart_export.txt")
```

```python
biomart_mapping = biomart_mapping.rename(columns = {"Gene stable ID": "token_name", "HGNC symbol": "gene_symbol"})
```

```python
merged_df = pd.merge(biomart_mapping, merged_df, on='token_name')
```

```python
merged_df = merged_df.drop('index', axis=1)
merged_df = merged_df.drop('Gene stable ID version', axis=1)
merged_df = merged_df.drop('token_name', axis=1)
```

```python
#merged_df
#merged_df[merged_df['gene_symbol'] == ""]
merged_df = merged_df.replace('', pd.NA).dropna()
```

```python
merged_df['gene_symbol'].nunique()
```

```python

merged_df = merged_df.groupby('gene_symbol').mean()

# If you want to keep the 'gene_symbol' column as a regular column instead of an index, reset the index
merged_df = merged_df.reset_index()
```

```python
import pandas as pd
import umap
import matplotlib.pyplot as plt

# UMAP transformation
reducer = umap.UMAP()
embedding = reducer.fit_transform(merged_df.drop("gene_symbol", axis=1))

```

```python
plt.figure(figsize=(12, 12))
plt.scatter(embedding[:, 0], embedding[:, 1], s=2)
plt.title("UMAP Visualization")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()
```

```python
embedding_df = pd.DataFrame(embedding)
embedding_df['gene_symbol'] = merged_df['gene_symbol']
embedding_df = embedding_df.rename(columns={1: 'UMAP 1', 0: 'UMAP 0'})
embedding_df.head()
```

```python
#reorder columns
cols = embedding_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
embedding_df = embedding_df[cols]
```

```python
embedding_df.to_csv("gene_symbol_summarized_UMAP_Geneformer.csv")
```
